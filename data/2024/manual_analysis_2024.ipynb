```python
# Read CSV file
df = pd.read_csv(r'C:\Users\a00556457\Repos\personal-time-dashboard-streamlit\data\2024\2024_outlook_data.csv')
```

```python
# selection of important values. 
filtered_df = df[
    (df['Subject'].str.contains('Φίλοι|Exercise|Οικογένεια|Erroneous Tasks|<3|Σπίτι|Χαζεύω|Ύπνος|Break|Other|Blyat|Εργασία|Διάβασμα', case=False, na=False)) |
    (df['Categories'] == 'Yellow category')
]
```

```python
filtered_df
```

```python
# Remove text in parentheses including the parentheses
filtered_df['Subject'] = filtered_df['Subject'].str.replace(r'\(.*?\)', '', regex=True)

# Optional: Strip leading/trailing whitespace
filtered_df['Subject'] = filtered_df['Subject'].str.strip()
```

```python
filtered_df
```

```python
# Combine Start Date and Start Time, End Date and End Time into datetime objects
filtered_df['Start DateTime'] = pd.to_datetime(filtered_df['Start Date'] + ' ' + df['Start Time'], format="%d/%m/%Y %H:%M:%S")
filtered_df['End DateTime'] = pd.to_datetime(filtered_df['End Date'] + ' ' + df['End Time'], format="%d/%m/%Y %H:%M:%S")

# Calculate time spent
filtered_df['Time Spent'] = filtered_df['End DateTime'] - filtered_df['Start DateTime']

# Convert the result to minutes
filtered_df['Time Spent (Minutes)'] = filtered_df['Time Spent'].dt.total_seconds() / 60
```

```python
filtered_df
```

```python
# Add Month and Day columns
filtered_df['Month'] = filtered_df['Start DateTime'].dt.strftime('%B')  # Month name
filtered_df['Day'] = filtered_df['Start DateTime'].dt.day
```

```python
filtered_df.to_csv('filtered_data.csv')
```
